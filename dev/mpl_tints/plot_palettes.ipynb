```python
import jm_utils.data.mpl_tints as mtints
```

```python
mtints.register_palette_from_color('master_cmap', 'mastercard')
mtints.register_palette_from_color('mtm_cmap', 'MTM')
mtints.register_palette_from_color('python_cmap', 'python')
mtints.register_palette_from_color('rolls_royce_cmap', 'rolls-royce')
mtints.register_palette_from_color('purple_cmap', 'purple')
mtints.register_palette_from_color('orange_cmap', 'orange')
```

```python
fig = mtints.plot_palettes(['master_cmap', 'mtm_cmap', 'python_cmap', 'rolls_royce_cmap', 'purple_cmap', 'orange_cmap'], n_colors=32)
```

```python
## Vamos a hacer el caso de crear una paleta de Python
py_lst = [mtints.PROGRAMMING_COLORS[key] for key in mtints.PROGRAMMING_COLORS.keys() if 'Python'.lower() in key]
# py_lst
mtints.register_mpl_palette('Python', py_lst, n_bins=len(py_lst))
# mtints.register_mpl_palette()
```

```python
## Vamos a hacer el caso de crear una paleta de MTM
mtm_lst = [mtints.FINANCIAL_COLORS[key] for key in mtints.FINANCIAL_COLORS.keys() if 'MTM' in key]
mtints.register_mpl_palette('Montemar', mtm_lst, n_bins=len(mtm_lst))
# mtints.register_mpl_palette()
```

```python
## Vamos a hacer el caso de crear una paleta de MTM
visa_lst = [mtints.FINANCIAL_COLORS[key] for key in mtints.FINANCIAL_COLORS.keys() if 'visa' in key]
mtints.register_mpl_palette('visa', mtm_lst, n_bins=len(visa_lst))
# mtints.register_mpl_palette()
```

```python
fig = mtints.plot_palettes(['Montemar', 'Python', 'visa'], continous=True)
```
